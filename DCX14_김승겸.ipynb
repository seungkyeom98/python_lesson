```python
# 문제 3
list1= [1,50,27,19,37,33,5,3,48]
even=0
odd=0
for i in list1:
    if i%2==0:
        even+=1
    else:
        odd+=1
print(f"짝수 {even}개, 홀수 {odd}개")
```

```python
# 문제 4
score_list=[['김미희',89,78,99],
            ['서대희',67,88,77],
            ['강예진',100,55,96]]

# DF로 변환
noName=[]
for i in range(len(score_list)):
    noName.append(score_list[i][1:])
noName

name=[]
for i in range(len(score_list)):
    name.append(score_list[i][0])
name

col=['Python','Java','DB']

scoreDF=pd.DataFrame(noName) #index=name,columns=col
scoreDF

# 평균 계산
scoreDF['평균'] =scoreDF.sum(axis=1)//len(scoreDF.columns)
scoreDF

# 과목별 평균
# print(f"{(scoreDF.sum()//len(scoreDF))[0]}")

#학점
print("================================================")
print('이름','Python','Java','DB','평균','학점',sep='\t')
print("================================================")
for i in range(len(scoreDF)):
    if scoreDF['평균'][i] >=90:
        abc='A'
    elif scoreDF['평균'][i] >=80:
        abc='B'
    elif scoreDF['평균'][i] >=70:
        abc='C'
    elif scoreDF['평균'][i] >=60:
        abc='D'
    else:
        abc='F'
    print(name[i],scoreDF.iloc[i,0],scoreDF.iloc[i,1],scoreDF.iloc[i,2],scoreDF.iloc[i,3],abc,sep="\t")
print("================================================")
print('평균',(scoreDF.sum()//len(scoreDF))[0],(scoreDF.sum()//len(scoreDF))[1],(scoreDF.sum()//len(scoreDF))[2],sep='\t')



```

```python
# 문제 5
movie = np.genfromtxt('lib/data/ratings.dat',delimiter='::',dtype=np.int64)

movieId= np.unique(movie[:,1])
np.sort(movieId)

movieMeanList=[]
for i in movieId:
    movieMean=movie[movie[:,1]==i][:,2].mean()
    movieMeanList.append([i,movieMean])

movieMeanList
```

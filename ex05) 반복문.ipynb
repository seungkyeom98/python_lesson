### 반복문
- 프로그램 내에서, 똑같은 명령을 일정 횟수만큼 반복하여, 수행하도록 제어하는 명령문
- while: 반복횟수가 명확하지 않을 때
- for: 반복횟수가 정확할 때
- break: 반복문을 나가는 기능.

### 파이썬 최고를 5번 출력하세요

```python
cnt = 1 #cnt는 count의 준말 변수명!!
        #반복횟수를 조절하기 위한 변수

while cnt <= 5:
    print('파이썬 최고')
    cnt+=1

# cnt가 6이 되는 순간, while문의 조건이 거짓이기 때문에
# 실행문장을 실행하지 않고, while문을 빠져나온다.
```

```python
# 무한루프
# i를 두번 누르면, 실행이 강제 정지.
n=1
while True:
    print(n)
    break
    n+=1
    #break
```

```python
cnt=1

while True:
    if cnt >5:
        break  # cnt변수가 5이상이면, 반복문을 빠져나가도록

    print('파이썬 최고')
    cnt += 1
```

```python
list2= [4,5,2,1,99,15,2,7,27]

max=0 # 모든 인덱스의 데이터가 음수라면? 그러므로 max를 0으로 초기화하면 안된다! 

max=list2[0] #이렇게 초기화 하면 된다.!


for i in list2:
    if i >max: # 기존 max값보다 더 큰수라면,
        max=i  # max에 i값을 넣어준다.

print(max)
```

### range() 함수
- 필요한 만큼의 숫자를 만들어주는 함수
- range(시작할 숫자, 종료할 숫자, 증가량) (끝값은 포함되지 않음.)
- (list[:3]과 같은 인덱스 슬라이싱처럼 마지막 인덱스는 포함안된다)

#### range(1,10,1) -> 1부터 9까지 1씩 증가
- 1,2,3,4,5,6,7,8,9

#### range(3,100,5) -> 3부터 99까지 5씩 증가
- 3,8,13,18, ... , 93,98

#### range(10,1,-1) -> 10부터 2까지 1씩 감소 (끝값은 포함하지 않기 때문이다)
- 10,9,8,7,6,5,4,3,2

```python
for i in range(1,10,1):
    print(i, end=' ')
```

```python
for i in range(3,100,5):
    print(i*10, end=' ')
```

#### range(인덱스 길이)

```python
# 파이썬 반복문(for)
for i in range(3):
    print('안녕')
```

```python
num = int(input("정수 입력 >> "))
list1 = []
print(f'{num}의 약수 : ',end='')
for i in range(1,num+1):
    if num%i==0:
        print(i,end=' ')
        list1.append(i)

print(f'\n{num}의 약수 : {list1}')
del list1[:]   #인덱스 전체 삭제
# del list1    #이건 list1이란 리스트 변수자체를 삭제.  2개 다르다.!
print(list1)
```

```python
list4= [1,2,3]
list5= [3,4,5,6]
list6= []
if( len(list4)>= len(list5)):
    length = len(list5)
    for i in range(length):
        list6.append(list4[i]+list5[i])
    for i in range(len(list4)-length):
        list6.append(list4[i+length])
else:
    length = len(list4)
    for i in range(length):
        list6.append(list4[i]+list5[i])
    for i in range(len(list5)-length):
        list6.append(list5[i+length])

print(list6)
```

```python
list4 = [1,2,3,4,5,6,7,8]
list5 = [3,4,5,6]
list6 = []

print('list4 :', list4)
print('list5 :', list5)

# 몇번 append 할건지?
# n이라는 변수에 넣어주자

# 두 리스트 길이의 차이만큼, 더 긴 리스트의 값이 들어간다.
# 두 리스트 길의 차이 n2라는 변수에 넣어주자



if len(list4) < len(list5): #list4의 길이가 작을 경우
    n = len(list4)
    n2 = len(list5) - len(list4)
else:
    n = len(list5)
    n2 = len(list4) - len(list5)

for i in range(n):
    list6.append(list4[i]+list5[i])

print('중간점검 :', list6)

# 뒷부분 더 긴 리스트에만 있는 값 append()
for i in range(n2):
    if len(list4) > len(list5):
        list6.append(list4[n+i])  # 4+0, 4+1, 4+2, 4+3
    else:
        list6.append(list5[n+i])

print('최종 :', list6)
```

```python
list4= [1,2,3,4,5,6,7,8]
list5= [3,4,5,6]
list6=[]

if len(list4)>len(list5):
    #앞
    for i in range(len(list5)):
        list6.append(list4[i]+list5[i])
    #뒤
    for i in range(len(list5),len(list4)):
        list6.append(list4[i])
else:
    #앞
    for i in range(len(list4)):    
        list6.append(list4[i]+list5[i])
    #뒤
    for i in range(len(list4),len(list5)):    
        list6.append(list5[i])
print(list6)
```

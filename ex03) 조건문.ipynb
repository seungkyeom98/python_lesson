## 조건문
- 상황에 따라 실행 흐름을 다르게 하는 문법
- 키워드 : if, elif, else
- 콜론(:)과 들여쓰기(tap) 문법이 중요!

```python
# 들여쓰기로 조건문의 실행문장을 구분한다.
if True:
    print('혁제씨는 잘생겼다.')

print('이 문장은 조건문 밖에있는 문장이라 항상 실행된다.')
```

```python
if False:
    print('혁제씨는 차은우 닮았다.')

print('이 문장은 조건문 밖에있는 문장이라 항상 실행된다.')
```

```python
# 간단한 if 조건문
money = 11000

if money >= 10000 : #조건식
    print('택시를 탄다.') #실행문장
    
```

```python
# 간단한 if 조건문
money = 11000

if money >= 10000 : #조건식
    print('택시를 탄다.') #실행문장
else:
    print('버스를 탄다.')
```

```python
num = int(input('정수 입력>>')) #문자열이니까 int()로 숫자형으로 바꾼다.
                                #반대로 문자 str()으로 바꿀수 있다.

if num%3 ==0 and num%5 ==0:
    print('3과 5의 배수입니다.')
else:
    print('3과 5의 배수가 아닙니다.')
```

```python
num1 = int(input('첫번째 정수 입력>>'))
num2 = int(input('두번째 정수 입력>>'))

if num1 > num2:
    print('첫번째 정수가 더 큽니다.')
elif num1<num2:
    print('두번째 정수가 더 큽니다.')
else:
    print('두 정수가 같습니다.')
```

```python
score = input('점수 입력 >>')

# 입력값이 숫자인지 먼저 확인
if score.isdigit():
    # 숫자일 경우에만 학점 계산 로직 실행
    if int(score)<=100 and int(score) >=0:
        if int(score)>= 90:
            print(f'{score}점은 A학점 입니다.')
        elif int(score)>=80:
            print(f'{score}점은 B학점 입니다.')
        elif int(score)>=70:
            print(f'{score}점은 C학점 입니다.')
        elif int(score)>=60:
            print(f'{score}점은 D학점 입니다.')
        else:
            print(f'{score}점은 F학점 입니다.')
    else:
        print('100~0점 사이 점수를 입력해주세요.')
else:
    print('숫자가 아닙니다. 100~0점 사이 점수를 입력해주세요.')
```

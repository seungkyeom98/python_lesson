### 스크롤 내리기
- 효율적으로 내리기 위해 창 최대화!

````python
import pandas as pd
````

````python
# 창 최대화
driver.maximize_window()
````

````python
# END, Page Down, Space bar, 방향키(아래)
# END가 가장 효율적!
body=driver.find_element(By.TAG_NAME, "body")
````

````python
body.send_keys(Keys.END)
````

````python
# 페이지 길이 가져오기!
driver.execute_script("return document.body.scrollHeight") # 스크롤 길이 확인
# 비동기로 인해 길이값이 제대로 업데이트가 이루어지지 않는 상황도 발생!
# 그래서 좀 더 쉽게 비교하는 방법!
````

````python
# 페이지 스크롤 내리기 알고리즘!
while True:
    # body= driver.find_element(By.TAG_NAME,"body")
    old = driver.page_source # 기존 페이지 소스( 문자열 자료형)

    #스크롤 하강
    body= driver.find_element(By.TAG_NAME, "body")
    body.send_keys(Keys.END)
    time.sleep(2)
    
    new = driver.page_source # 갱신된 페이지 소스

    if old == new: #같은지 확인
        print("스크롤 하강 완료")
        break
    else:
        pass # 코드 리뷰할때 이러면 편하다
````

````python
#요소가 완전히 로드된 상태에서 img 태그 접근
soup=bs(driver.page_source, "lxml") # 이게 파싱 하는것. 문자열을 html로 바꿈
imgs = soup.select("a > div.img > div >img.img-fluid")
````

````python
#반복문 진행현황 출력!
from tqdm import tqdm #타카돔 -> 반복문 현황 출력하는 도구

for img in tqdm(imgs):
    imgLink.append(img["src"])
````

````python
# 링크만 수집하기!
imgLink= [img["src"] for img in imgs] # 리스트 내포, 컴프리헨션
````

### 실습! 코드 하나의 셀에 합치기!
1. 폴더 생성 자동화
2. 스크롤 하강
3. 링크 수집
4. 링크 기반으로 다운로드 -> tqdm 활용!
5. 자원반납

````python
from selenium import webdriver as wb # 테스트 브라우저 관리 도구
from selenium.webdriver.common.by import By #경로 지정방식
from selenium.webdriver.common.keys import Keys #키보드 입력값

from bs4 import BeautifulSoup as bs # 파싱 도구
import time # 시간지연 딜레이 라이브러리
from tqdm import tqdm #타카돔 -> 반복문 현황 출력하는 도구
from urllib.request import urlretrieve # url 기반으로 다운로드 할 수 있는 도구!

#1폴더 생성
try:
    os.mkdir("C:\\Users\\smhrd1\\Desktop\\포켓몬")
except:
    print("폴더가 이미 존재함")

#2스크롤 하강
        # 브라우저 세팅
driver = wb.Chrome()
url = "https://www.pokemonkorea.co.kr/pokedex#pokedex_1"
driver.get(url)
driver.maximize_window()
body=driver.find_element(By.TAG_NAME, "body")
body.send_keys(Keys.END)
        # 페이지 스크롤 내리기 알고리즘!
while True:
    # body= driver.find_element(By.TAG_NAME,"body")
    old = driver.page_source # 기존 페이지 소스( 문자열 자료형)

    #스크롤 하강
    body= driver.find_element(By.TAG_NAME, "body")
    body.send_keys(Keys.END)
    time.sleep(2)
    
    new = driver.page_source # 갱신된 페이지 소스

    if old == new: #같은지 확인
        print("스크롤 하강 완료")
        break
    else:
        pass # 코드 리뷰할때 이러면 편하다
        
#3링크만 수집하기!
    #요소가 완전히 로드된 상태에서 img 태그 접근
soup=bs(driver.page_source, "lxml") # 이게 파싱 하는것. 문자열을 html로 바꿈
imgs = soup.select("a > div.img > div >img.img-fluid")
    #img 태그의 src속성인 이미지링크만 수집! 
imgLink= [img["src"] for img in tqdm(imgs)]

#4링크로 이미지 다운로드 -> tqdm활용
list(map(lambda a,b:urlretrieve(a,f"C:\\Users\\smhrd1\\Desktop\\포켓몬\\포켓몬{b+1}.jpg"), tqdm(imgLink),range(len(imgLink))))

#5자원반납
driver.quit()
````

```python
matrix=[[1,2,3],[4,5,6],[7,8,9]]
```

```python
#list의 요소 변경 및 추가
list5= [0,1,2]
print('변경 전', list5)

#[0,1,5]
list5[-1] =5
print('변경 후', list5)
print(list5[-1])

#[0,1,5,6]
#append()
#list의 요소를 추가
list5.append(6)
print('변경 후', list5)
```

```python
# 슬라이싱으로 요소 변경
list5 = [0,1,2,3,4,5,6]
print('변경 전', list5)

list5[2:5]= [4863] 
print('변경 후', list5)
# 슬라이싱한 값은 자료형이 리스트이므로, 대입도 자료형 리스트로 맞춰줘야한다.

list5[2:5]= ['a','헷',9,9,9,9] 
print('변경 후', list5)
```

```python
array=[1,2,3,4,5]

array[4]=7
print(array,'\n')

array[1]=['a','b','c'] #인덱스 1번 '요소'에 접근한다.
print(array,'\n')      #그래서 1번 인덱스 요소에 그 대입된 값으로 어떤 자료형이던간에 대입되고

array[1:2]=['a','b','c'] # '리스트'1~2사이 '리스트'에 접근한다. 
print(array,'\n')        # 슬라이싱은 리스트 1~2사이 인덱스가 대입된 리스트로 바뀌는 걸
                         # 슬라이싱한 리스트에 대입이다!

array[-2:]=['d','e','f','g']
print(array,'\n')
```

```python
fruits = ['사과','포도',['수박','멜론'],'오렌지']

apple = fruits[0] #인덱스은 해당 요소의 자료의 type이고!
                  #슬라이싱은 자료의 형태가 리스트type이다!!!!!!!!!!!!!!!! 
                  #헷갈리지 말자. 그 값이 '해당 요소의 type'의 값인지 '리스트 type'의 값인지!
melon = fruits[2][1]
orange = fruits[3]
```

```python
print(apple)
print(melon)
print(orange)
```

```python
choice_1st= [] # 과일을 담아줄 빈 리스트 정의
choice_1st.append(apple)
choice_1st.append(melon)
choice_1st.append(orange)

choice_1st
```

```python
# 한번에 값을 담아주는 함수
choice_1st2 = []

#extend: 리스트 안에 여러개의 값을 한번에 추가할 때
choice_1st2.extend([apple,melon,orange]) # list의 형태로 넣어준다.

choice_1st2
```

위 리스트의 길이(=크기=length)는 3이다!!! 

### 리스트의 다양한 함수들

```python
#리스트의 길이(요소 개수)를 반환하는 함수
#len
lst = [2,34,2,6,32,34,2,34,24,2,6,6,34234,1,342,43,4]
print(len(lst))
```

```python
#lst = list의 준말 변수명.
```

```python
#특정 값의 인덱스 반환
#index()
#list.index(찾고자하는 값)

lst.index(6)  #가장 첫번째 요소값이 6인 인덱스 번호를 알려준다!
```

```python
#값을 추가할 때, 맨 끝이 아닌 내가 원하는 위치 추가
#insert
#list.insert(인덱스위치, 값)

list5 = [0,1,2,3,4]
print('변경 전: ',list5)

list5.insert(2,125)
print('변경 후: ',list5) #원래 있던 요소를 다음 인덱스로 밀어버리고 그자리에 삽입된다.

list5.insert(2,[1,3,5])
print('변경 후: ',list5)
```

```python
#list 값 삭제
list7 = [33,44,55,66,77,44,44,44,44,44]
print('삭제 전: ', list7)

# 인덱스 번호로 삭제   delete내장함수의 키워드 : del
# del
del list7[3]
print('삭제 후: ', list7)

# 값으로 삭제
# remove()
# list.remove(삭제할 값)
list7.remove(44)
print('44 삭제 후: ', list7)
```

```python
# 슬라이싱을 활용해서 삭제
list7 = [33,44,55,66,77,44,44,44,44,44]
print('삭제 전: ', list7)

# del
del list7[6:]
print('삭제 후: ', list7)
```

```python
# 오름차순 정렬
# sort()
# list.sort()

list8 = [6,3,5,8,2,1]
print('정렬 전: ',list8)

list8.sort()
print('정렬 후: ',list8)
```

```python
# 역순 
# 리버스 내장 함수 = reverse()

list8 = [6,3,5,8,2,1]
print('정렬 전: ',list8)

list8.reverse()
print('역순 후: ',list8)
```

```python
list8 = [6,3,5,8,2,1]
print('정렬 전: ',list8)

list8.sort()
print('정렬 후: ',list8)

list8.reverse()
print('역순 정렬 후: ',list8)
```

### 비교 연산자
- 조건식 사용시 불리언 값을 반환(True, False)

```python
# 조건식
3>5
```

```python
a=3
b=7

print(a > b)
print(a >= b)
print(a == b) # =는 대입연산자, ==는 비교연산자
print(a != b,'\n')

#논리 연산자 not: 논리값을 뒤집음
print(not a > b) # a < b
print(not a >= b)
print(not a == b)
print(not a != b)
```

### 논리연산자 and, or

- or : 어느 하나라도 True라면, True

```python
5>3 and 7>2
```

```python
5>3 or 1>3
```

### 멤버 연산자 in

```python
s = '콩국수에 들어갈건 소금뿐'

# 조건식
'소금' in s
# '소금'이라는 문자가 s 안에 있는가?
```

### 삼항연산자
- 조건문의 단순화
- a if 조건식 else b

```python
# 삼항연산자 예시
java_score = int(input('자바 예상 점수 입력: '))

'합격' if java_score >= 60 else '불합격'
# 합격이다. 만약에, java점수가 60 이상이라면, 그렇지 않다면, 불합격이다.
```

```python
a = int(input("정수입력>>"))
b = int(input("정수입력>>"))

print(f'a: {a}') if a>b else print(f'b: {b}')
```

```python
a= int(input('정수 입력 >>'))

print(f'{a}는(은) 짝수입니다.') if a%2==0 else  print(f'{a}는(은) 홀수입니다.')
```

```python
a =  int (input('정수 입력>>'))

# 홀수, 짝수 여브를 저장할 변수
eo = '홀수' if a%2==1 else '짝수'

print(f'{a}는(은) {eo}입니다.')

print(a,'는(은) ',eo,'입니다.',sep='')

print(str(a)+'는(은)', eo+'입니다.')
```

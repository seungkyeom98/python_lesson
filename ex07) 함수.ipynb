### 함수 사용 목적
- 반복적인 프로그래밍을 피할 수 있다.
- 모듈화로 인해 코드의 가독성이 좋아진다.
- 프로그램에 문제가 발생하거나, 기능의 변경이 필요할 때, 손쉽게 유지보수가 가능하다.

```python
num1 = int(input('첫번째 정수 입력>>'))
num2 = int(input('두번째 정수 입력>>'))

# 빼기 기능을 하는 함수를 호출 (두 인자값을 입력하여 호출)

result = number_sub(num1, num2)
print(result)
```

```python
# number_sub 라는 함수를 정의
def number_sub(n1,n2):  # 함수를 정의, 매개변수로는 n1, n2를 사용한다.
    result= n1-n2       # 실행문장, result라는 변수에, 매개변수끼리 뺀 값을 저장한다.
    return result       # result변수의 값을 반환
```

```python
# 방법1
# 함수를 정의

def s_replace(str):
    result='' # result라는 문자열형 변수 선언
    for i in str:
        if i != 'ㅋ': # i가 'ㅋ'이 아니라면~
            result= result + i # result라는 변수에 i를 더해준다.(원래값 오른쪽에 탕탕탕 붙여준다.)
           #result= i +result하면, 원래 값에 왼쪽으로 쭉 추가 시켜줄수 있다. (역순으로 추가하는 방법임!)
    return result
```

```python
# replace() 함수
# str.replace(바뀔문자, 바꿀문자)

str1 = '카ㅓ카ㅓㅠ카ㅓㅠ커ㅏ큐ㅓㅏㅋㅇㅋㅇㅇㄴㅋ'
print('바꾸기 전 :', str1)

str1.replace('ㅋ','') # 원본에 영향을 미치지 않는다. 
                      #그러므로, 재할당하거나 새로 변수에 할당(=저장)하면 된다.
```

```python
def divisor(n1):
    result=''
    for i in range(1,n1+1): # n1은 시퀀스가 아니니까, range()함수를 써야한다!
        if n1%i==0:
            result= result+str(i)+' '
    return result
```

```python
print(divisor(100)) #반환이 없는 함수는 print안에 넣으면, none이 나온다.
```

```python
def divisor(n1):
    for i in range(1,n1+1): # n1은 시퀀스가 아니니까, range()함수를 써야한다!
        if n1%i==0:         #끝값은 포함되지 않으니까, n+1을 한다~
            print(i, end=' ')
    
```

### 독스트링

```python
def divisor(n1: int): #매개변수에 int형(=정수형) 데이터(=자료=값)을 넣어라
    '''
    약수를 구해서 출력해주는 함수
    여러줄 입력 가능
    '''
    for i in range(1,n1+1): 
        if n1%i==0:         
            print(i, end=' ')
    
```

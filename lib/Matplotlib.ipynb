# Matplotlib

#### Mathdatas plot library
- 메타데이터를 도표나 그래프(시각화)로 출력시키는 라이브러리
- 메타데이터 : 일정한 규칙에 따라 구조화, 표준화된 데이터
- 분석 결과를 수치만으로 설명하는 것보다 더 직관적으로 이해 가능
- 타인에게 분석 결과를 전달하기에 효과적

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

#### 플롯 차트(plot chart) : 선 그래프

```python
y = [2,4,6]

plt.plot(y) # 선 그래프 만들어주는 함수

plt.show() # 차트를 화면에 출력. (주피터 노트북에서는 생략해도 출력)
```

```python
# x,y 값 지정하여 선 그래프 그려보기
x =[1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x,y)

plt.show()
```

```python
# x,y 값 지정하여 선 그래프 그려보기
x =[1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x,y, ls= '-.', lw = 5, c = 'violet', marker = 'p', mfc = 'blue', mec = 'green', ms = 12, mew = 1.5)
# ls : 선 스타일
# lw : 선 굵기
# c : 선 색깔
# marker : 마커의 종류 (데이터(=자료=값)의 위치를 마킹해주는 것.)
# mfc : 마커 색깔
# mec : 마커의 테두리 색깔
# ms : 마커의 사이즈(크기)
# mew : 마커의 테두리 선 굵기 (너비)

plt.show()
```

### 그리드
- 차트의 선 위치들을 잘 보여주기 위해 차트 중간에 각자 선을 그어주는 것

```python
# x,y 값 지정하여 선 그래프 그려보기
x = np.arange(7)
y = [1,4,6,8,9,5,3]

plt.plot(x,y, ls= '-.', lw = 5, c = 'violet', marker = 'p', mfc = 'blue', mec = 'green', ms = 12, mew = 1.5)
plt.grid()

plt.xticks([0,3,6])
plt.yticks([1,5,9])

plt.xlabel('X axis') # x축의 제목, axis=에이시스
plt.ylabel('Y axis') # Y축의 제목
plt.title('X-Y graph') # 그래프 전체 제목


plt.show()
```

### 틱(tick) 설정
- 플롯이나 차트에서, 축 상의 위치 표시 지점 : 틱(tick)
- 틱 위에 써진 숫자, 문자 : 틱 라벨(tick label)
- 자동으로 설정, 수동 설정도 가능

```python
x = np.arange(4)

y1 = [1,2,3,4]
y2 = [1,4,6,8]
y3 = [3,6,9,12]

plt.plot(x,y1, label= 'y1')
plt.plot(x,y2, label= 'y2')
plt.plot(x,y3, label= 'y3')

# 범례 설정 legend
plt.legend(loc= 'upper right',prop = {'size':20})
# loc : 범례의 위치를 표시 (upper, center, lower) + ( left, center, right) # 정중앙은 center (center center가 아니다)

plt.show()
```

### 바 차트(bar chart) : 막대그래프

```python
x = np.arange(6)
y = [80,85,70,60,50,90]

plt.bar(x,y, color = ['orange','aqua', 'pink'], edgecolor = 'blue',lw = 4, alpha = 0.5, width = 0.7)

# color : 바 내부 색상
# edgecolor : 바 테두리 색상
# linewidth(lw) : 테두리 너비
# alpha : 투명도 설정
# width : 막대 너비


plt.show()
```

```python
n = 8
X,Y = np.mgrid[0:n,0:n]
T = np.arctan2(Y-n/2.0, X-n/2.0)
R = 10+np.sqrt((Y-n/2.0)**2+(X-n/2.0)**2)
U,V = R*np.cos(T), R*np.sin(T)

plt.figure(figsize=(10,8))
plt.quiver(X,Y,U,V,R, alpha=.5)
plt.quiver(X,Y,U,V, edgecolor='k', facecolor='None', linewidth=.5)

plt.xlim(-1,n), plt.xticks([])
plt.ylim(-1,n), plt.yticks([])

# savefig('../figures/quiver_ex.png',dpi=48)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # 주피터 노트북에서 필요할 수 있으므로 남겨둡니다.

# Figure와 subplot 설정
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

# X, Y, Z 데이터 설정
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# 3D 표면 플롯
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.hot)
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.cm.hot)
ax.set_zlim(-2,2)

# 플롯 표시
plt.show()
```

주피터 노트북에서 위 코드를 실행할 때 <Figure size 1000x1000 with 0 Axes>와 같은 메시지가 나타나는 것은 Axes3D 객체가 제대로 초기화되지 않았기 때문일 가능성이 있습니다. 이를 해결하려면, 다음과 같은 사항을 확인해보세요.

matplotlib의 3D 기능을 사용할 때 Axes3D 모듈을 직접 가져오는 대신 fig.add_subplot()과 projection='3d' 옵션을 사용하는 방법이 있습니다.
from mpl_toolkits.mplot3d import Axes3D는 이제 ax = fig.add_subplot(111, projection='3d')로 대체할 수 있습니다.
아래는 수정된 코드입니다. 이를 주피터 노트북에서 실행해보세요:

```python
eqs = []
eqs.append((r"$W^{3\beta}_{\delta_1 \rho_1 \sigma_2} = U^{3\beta}_{\delta_1 \rho_1} + \frac{1}{8 \pi 2} \int^{\alpha_2}_{\alpha_2} d \alpha^\prime_2 \left[\frac{ U^{2\beta}_{\delta_1 \rho_1} - \alpha^\prime_2U^{1\beta}_{\rho_1 \sigma_2} }{U^{0\beta}_{\rho_1 \sigma_2}}\right]$"))
eqs.append((r"$\frac{d\rho}{d t} + \rho \vec{v}\cdot\nabla\vec{v} = -\nabla p + \mu\nabla^2 \vec{v} + \rho \vec{g}$"))
eqs.append((r"$\int_{-\infty}^\infty e^{-x^2}dx=\sqrt{\pi}$"))
eqs.append((r"$E = mc^2 = \sqrt{{m_0}^2c^4 + p^2c^2}$"))
eqs.append((r"$F_G = G\frac{m_1m_2}{r^2}$"))

plt.figure(figsize=(10,10))

for i in range(24):
    index = np.random.randint(0,len(eqs))
    eq = eqs[index]
    size = np.random.uniform(12,32)
    x,y = np.random.uniform(0,1,2)
    alpha = np.random.uniform(0.25,.75)
    plt.text(x, y, eq, ha='center', va='center', color="#11557c", alpha=alpha,
             transform=plt.gca().transAxes, fontsize=size, clip_on=True)

plt.xticks([]), plt.yticks([])
# savefig('../figures/text_ex.png',dpi=48)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x,y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)

plt.figure(figsize=(10,8))

plt.contourf(X, Y, f(X,Y), 8, alpha=.75, cmap=plt.cm.hot)
C = plt.contour(X, Y, f(X,Y), 8, colors='black', linewidth=.5)
plt.clabel(C, inline=1, fontsize=10)

plt.xticks([]), plt.yticks([])
# savefig('../figures/contour_ex.png',dpi=48)
plt.show()
```

### 4. 아이디가 1인 사용자가 매긴 영화의 평점들의 평균

```python
# 아이디가 1인 사용자

# 조건 -> 불리언 인덱싱
# id ==1

movie[:,0]==1 #이게 조건식.
```

```python
# 불리언 인덱싱
# id가 1인 영화 데이터를 id_1이라는 변수에 저장

id_1 =movie[movie[:,0]==1]

id_1
```

```python
# id가 1인 사용자의 평점들

id_1_r=id_1[:,2]

id_1_r
```

```python
id_1_r.mean()
```

```python
# 중간 변수로 만드는 과정 없이

movie[movie[:,0]==1][:,2].mean() # = id_1[:,2].mean() 
```

### 5. 각 사용자가 매긴 평점들의 평균
- 리스트에 담기

```python
movie
```

```python
# 데이터의 유효성 검사( id가 정말 6040개 인지)

# 중복값을 제거해주는 (유일한 값들만 구해주는) 함수
# np.unique()

user_id = np.unique(movie[:,0]) # id 값들로 이루어진 배열을 unique함수에 넣음

user_id
```

```python
len(user_id)
# id는 총 6040개
# 사용자는 6040명
```

```python
# 오름차순 sort()

asdf = np.array([1,24,142,1,1,34134,14,141])

np.sort(asdf)
```

```python
np.sort(user_id) # nd배열은 numpy라이브러리 안에만 있는 자료형이라 np.sort를 써야한다. 그냥 sort함수는 안먹힌다.
```

힌트
- 1번 사용자가 매긴 평점들의 평균 구했다.
- 2번 사용자도 같은 방식
- 6040번 사용자도 같은 방식
- for문
- list에 담기

```python
# 비어있는 리스트를 정의, 각 사용자의 id의 평점의 평균
user_mean_list=[]

for i in user_id:
    user_mean = movie[movie[:,0]==i][:,2].mean()
    #사용자 별로 영화 평점의 평균을 user_mean이라는 변수에 넣어둠
    user_mean_list.append([i,user_mean])

user_mean_list
```

### 6. 각 사용자별 평균 평점이 4점 이상인 사용자만 구하기

```python
user_mean_arr = np.array(user_mean_list)

user_mean_arr
```

```python
# set_printoptions : numpy 배열 출력시 출력 형태 설정 함수
# suppress : 소수점 자리수 고정 출력
# precision : 소수점 자리수 

np.set_printoptions(suppress=True, precision=2)

user_mean_arr
```

```python
# 4점 이상인 데이터
# astype(int)
# 원본에 영향을 안 미침.

user_mean_arr[user_mean_arr[:,1]>=4][:,0].astype(int)
```

```python
user_mean_arr[user_mean_arr[:,1]>=4][:,0]
```

```python
movie_id = np.unique(movie[:,1])

movie_id
```

```python
# 비어있는 리스트를 정의, 각 사용자의 id의 평점의 평균
movie_mean_list=[]

for i in movie_id:
    movie_mean = movie[movie[:,1]==i][:,2].mean()
    #사용자 별로 영화 평점의 평균을 user_mean이라는 변수에 넣어둠
    movie_mean_list.append([i,movie_mean])

movie_mean_list
```

```python
movie_mean_arr = np.array(movie_mean_list)

movie_mean_arr
```
